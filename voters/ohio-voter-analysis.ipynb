```python
import pandas as pd
district1 = pd.read_csv("1_22_sample.csv")
district1.head()
```

```python
district2 = pd.read_csv("23_44_sample.csv")
district3 = pd.read_csv("45_66_sample.csv")
district4 = pd.read_csv("67_88_sample.csv")

voters = pd.concat([district1, district2, district3, district4])
voters.head()
```

```python
print(district1.shape)
print(district2.shape)
print(district3.shape)
print(district4.shape)
print(voters.shape)
```

```python
voters = voters.sample(frac=1)
cutoff = round(voters.shape[0] * 0.8)
cutoff
```

```python
train = voters[:cutoff]
test = voters[cutoff:]
train.head()
```

```python
from sklearn.model_selection import train_test_split

columns = voters.columns[np.arange(85, 99)]
sample = voters
all_X = sample[columns].fillna(0).astype("bool")
all_y = sample['GENERAL-11/08/2016'].fillna(0).astype("bool")

train_X, test_X, train_y, test_y = train_test_split(all_X, all_y, test_size=0.20)
train_X.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
lr = LogisticRegression()
lr.fit(train_X, train_y)
predictions = lr.predict(test_X)
accuracy = accuracy_score(test_y, predictions)
accuracy
```

```python
from datetime import datetime, timedelta

def date_to_days(date):
    start = datetime(year=2000, month=1, day=1)
    date = datetime.strptime(date, "%Y-%m-%d")
    diff = start - date
    return diff.days   

date_features = [
 'DATE_OF_BIRTH',
 'REGISTRATION_DATE'
]
float_cols = ["float_dob", "float_reg"]

sample = voters
dates = sample[date_features]
dates["float_dob"] = dates['DATE_OF_BIRTH'].apply(date_to_days)
dates["float_reg"] = dates['REGISTRATION_DATE'].apply(date_to_days)
dates.head()

all_X = dates[float_cols]
all_y = sample['GENERAL-11/08/2016'].fillna(0).astype("bool")

train_X, test_X, train_y, test_y = train_test_split(all_X, all_y, test_size=0.20)
train_X.head()
```

```python
lr = LogisticRegression()
lr.fit(train_X, train_y)
predictions = lr.predict(test_X)
accuracy = accuracy_score(test_y, predictions)
accuracy
```

```python
location_features = [
 'CONGRESSIONAL_DISTRICT',
 'COURT_OF_APPEALS',
 'LIBRARY',
 'STATE_BOARD_OF_EDUCATION',
 'STATE_REPRESENTATIVE_DISTRICT',
 'STATE_SENATE_DISTRICT'
]

locations = sample[location_features]
location_dummies = pd.get_dummies(locations, dummy_na = True)
location_dummies.head()
```

```python
congressional_dummies = pd.get_dummies(locations["CONGRESSIONAL_DISTRICT"], prefix = "district")
congressional_dummies.head()
```

```python
court_dummies = pd.get_dummies(locations["COURT_OF_APPEALS"], prefix = "court")
court_dummies.head()
```

```python
sum(pd.notnull(locations["LIBRARY"]))
```

```python
board_dummies = pd.get_dummies(locations["STATE_BOARD_OF_EDUCATION"], prefix = "board")
board_dummies.head()
```

```python
representative_dummies = pd.get_dummies(locations["STATE_REPRESENTATIVE_DISTRICT"], prefix = "representative")
representative_dummies.head()
```

```python
senate_dummies = pd.get_dummies(locations["STATE_SENATE_DISTRICT"], prefix = "senate")
senate_dummies.head()
```

```python
all_dummies = pd.concat(
    [congressional_dummies, court_dummies, representative_dummies, representative_dummies, senate_dummies], axis=1
)
all_dummies.head()
```

```python
all_X = all_dummies
all_y = sample['GENERAL-11/08/2016'].fillna(0).astype("bool")

train_X, test_X, train_y, test_y = train_test_split(all_X, all_y, test_size=0.20)

lr = LogisticRegression()
lr.fit(train_X, train_y)
predictions = lr.predict(test_X)
accuracy = accuracy_score(test_y, predictions)
accuracy
```
